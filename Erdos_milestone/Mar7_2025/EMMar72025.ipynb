We need to determine with what contextual parameters should we train the model. More economical knowledge and further processing might be needed for this sake. 

1, Our long time goal is the train a supervised model to predict the change of stock index with macroeconomic data as parameters. 

2, Since what we care about is the impact of macroeconomic data, the target of training will be the change of the stock index (instead of the actual index). As an example of proportional change of stock index, the change of close index on 2018-01-01 is 
$$\frac{\text{Close index on 2018-01-01}-\text{Close index on 2017-12-31}}{\text{Close index on 2017-12-31}}\;.$$

3, The parameter we will care about should be contextual. For instance, instead of the actual gold price on a certain day, we will use the proportional change of gold price on a certain day. However, certain values are contextual on its own, for instance, the interest rate of a certain day. 

4, The data we have is a time-series data by nature. However, we will also consider training (part of) the model by contextual parameter regardless of time: for instance, regardless of time, the proportional change in stock index when gold is changing in a certain way should be similar to that when gold is changing in a similar way. And we can use models like kernel regression to approach this. 

The time series training can be done with the resources themselves, for instance predicting the price of gold, which we can feed into our non-time-series model to predict stock index. Furthermore, we might be able to train a time-series model on the error of our non-time-series model to patch in any periodical changes.
