It is clear that one sic might have more than one naics, for instance sic 119. We then need to find a way to assign each sic to a list (or a set) of naics's. 

```python
df_NAICS_map_SIC.loc[0,"SIC"]
```

```python
set(df_NAICS_map_SIC.loc[df_NAICS_map_SIC.SIC=="119"].NAICS)
```

```python
dict_SIC_NAICS={}
for ind in range(2105):
    dict_SIC_NAICS[df_NAICS_map_SIC.loc[ind,"SIC"]]=set(df_NAICS_map_SIC.loc[df_NAICS_map_SIC.SIC==df_NAICS_map_SIC.loc[ind,"SIC"]].NAICS)
    
```

```python
dict_SIC_NAICS
```

There might be SIC values that was not assigned with a NAICS value. 

```python
lst_SIC_no_NAICS=[]

for sic in df_com_CIK_SIC.sic: 
    if not np.isnan(sic):
        if str(int(sic)) not in dict_SIC_NAICS: 
            lst_SIC_no_NAICS.append(sic)
```

```python
set(lst_SIC_no_NAICS) # a set containing all SIC that could not be assigned to a NAICS
```
